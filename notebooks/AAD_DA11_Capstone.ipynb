```python
def age_clean(age):
    if age =="Teen":
        return "17"
    if age =="Adult":
        return "18"
    if age =="Child":
        return "12"
    if age =="Minor":
        return "12"
    else:
        return(age)
```

```python
school_shootersdf['age2'] = school_shootersdf['Age'].apply(lambda x: age_clean(x))
```

```python
school_shootersdf['age2'] = pd.to_numeric(school_shootersdf['age2'])
```

```python
def ages(age): 
    if age <=12:
        return "Child"
    if  13<= age <=17:
        return "Teen"
    if age >=18:
        return "Adult"  
    if type(age)==str:
        return
```

```python
school_shootersdf['age_category'] = school_shootersdf['age2'].apply(lambda x: ages(x))
school_shootersdf
```
